```python
# 라이브러리 불러오기
import urllib.request 
import json 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

%config InlineBackend.figure_format='retina'

# 한글 쓰기
plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
```

```python
path = '추석교통량(최종).csv'

traffic = pd.read_csv(path)

traffic.head()
```

```python
traffic.rename(columns = {'stdHour':'시간대', 'stdYear':'연도','trfl':'교통량'},inplace = True)
traffic.drop('연도', axis = 1, inplace = True)
traffic
```

```python
path = '추석기온(최종).csv'

air = pd.read_csv(path)

air.drop('Unnamed: 0', axis = 1, inplace = True)
air
```

```python
air.drop(['지역','강수량'],axis = 1 , inplace = True)
air
```

```python
path = '통행시간(최종).csv'

min_ = pd.read_csv(path)

min_.head()
```

```python
min_.drop(['Unnamed: 0','시간'], axis = 1 , inplace = True)

min_
```

```python
fin = pd.concat([air,traffic, min_], join = 'inner', axis = 1)
fin=fin[['연도','시간대','출발영업소','도착영업소','온도','풍속','시정','교통량','통행시간(분)']]
fin.head()
```

```python
fin.to_csv('추석 연휴 시간대 예측.csv')
```

```python
fin.info()
```

```python
fin['통행시간(분)']= fin['통행시간(분)'].astype('int')
```

```python
import datetime
```

```python
def min_to_time(row):
    return str(datetime.timedelta(minutes = row))


fin['통행시간(분)']= fin['통행시간(분)'].apply(min_to_time)
```
